```python
from collections import deque

import numpy as np
```

```python
def parse_input(path):
    with open(path) as file_in:
        data = file_in.read().splitlines()

    splitters = set()
    for i, row in enumerate(data):
        for j, elem in enumerate(row):
            if elem == 'S':
                start = (i, j)
            if elem == '^':
                splitters.add((i, j))

    return start, splitters, len(data)
```

```python
def get_next_splitter(pos, splitters, n_rows, part):
    start = pos
    pos = pos[0]+1, pos[1]
    if pos[0] == n_rows-1:
        return None, start

    while pos not in splitters:
        pos = pos[0]+1, pos[1]
        if pos[0] == n_rows-1:
            return None, start

    return pos, start
```

```python
def main1(input_path):
    start, splitters, n_rows = parse_input(input_path)

    splitters_reached = set()
    queue = deque()
    queue.append(start)

    while queue:
        pos, __ = get_next_splitter(queue.popleft(), splitters, n_rows, part=1)
        if pos is not None and pos not in splitters_reached:
            splitters_reached.add(pos)
            queue.append((pos[0], pos[1]-1))
            queue.append((pos[0], pos[1]+1))

    return len(splitters_reached)
```

```python
assert main1("example.txt") == 21
main1("input.txt")
```

```python
def main1(input):
    with open(input) as file_in:
        input_str = file_in.read().splitlines()

    data = []
    for i, row in enumerate(input_str):
        row = row.split()
        data.append(row)

    n_operations = len(data[0])
    n_numbers_per_ops = len(data)-1

    total = 0
    for i in range(n_operations):
        numbers = [data[j][i] for j in range(n_numbers_per_ops)]
        symbol = data[-1][i]
        ops = f"{symbol}".join(numbers)
        total += eval(ops)

    return total
```

```python
def main2(input):
    with open(input) as file_in:
        input_str = file_in.read().splitlines()

    n_operations = len(input_str[0].split())
    n_char_per_operation = [0] * n_operations

    for row in input_str:
        numbers = row.split()
        for i in range(n_operations):
            n_char_per_operation[i] = max(n_char_per_operation[i], len(numbers[i]))

    data = []
    for row in input_str:
        row_parsed = []
        pointer = 0
        for n_char_col in n_char_per_operation:
            row_parsed.append(row[pointer:pointer+n_char_col])
            pointer += n_char_col + 1
        data.append(row_parsed)

    data[-1] = ["".join(s.split()) for s in data[-1]]

    total = 0
    for ops_idx in range(n_operations):
        symbol = data[-1][ops_idx]

        numbers = [data[i][ops_idx] for i in range(len(data)-1)]
        numbers_top_to_bottom = []
        for col in range(n_char_per_operation[ops_idx]):
            col_result = []
            for number in numbers:
                col_result.append(number[col])
            col_result = "".join([s for s in col_result if s != ' '])
            numbers_top_to_bottom.append(col_result)

        ops = f"{symbol}".join(numbers_top_to_bottom)
        ops_result = eval(ops)
        total += ops_result

    return total
```

```python
assert main1("example.txt") == 4277556
main1("input.txt")
```

```python
assert main2("example.txt") == 3263827
main2("input.txt")
```

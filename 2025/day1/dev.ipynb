```python
def apply_rotations(start_pos, rotations, part):

    pos = start_pos

    res_part1 = 0
    res_part2 = 0
    for rot in rotations:

        way = rot[0]
        distance = int(rot[1:])
        if way == "R":
            new_pos = (pos + distance) % 100
            n_turns = (pos + distance) // 100
        elif way == "L":
            new_pos = (pos - distance) % 100
            n_turns = abs((pos - distance) // 100)
            if pos == 0:
                n_turns -= 1
            if new_pos == 0:
                n_turns += 1
        res_part2 += n_turns

        # if part == 2:
        #     print(pos, rot, new_pos, n_turns)
        pos = new_pos
        
        if pos == 0:
            res_part1 += 1

    if part == 1:
        return res_part1
    else:
        return res_part2
```

```python
def main(input, part):

    with open(input) as file_in:
        rotations = file_in.read().splitlines()
    
    return apply_rotations(50, rotations, part)
```

```python
assert main("example.txt", part=1) == 3
main("input.txt", part=1)
```

```python
def parse_input(input):
    grid = []
    with open(input) as file_in:
        grid_str = file_in.read().splitlines()
        for row in grid_str:
            grid.append(list(map(int, row)))

    return grid
```

```python
def main(input, part):

    n_bat_per_row = 2 if part == 1 else 12

    grid = parse_input(input)

    total_joltage = 0
    for row in grid:
        row_largest_joltage = ""
        index = -1
        for n in range(n_bat_per_row, 0, -1):
            row_sub = row[index+1:len(row)-n+1]
            max_row = max(row_sub)
            row_largest_joltage += str(max_row)
            index += row_sub.index(max_row) + 1
        total_joltage += int(row_largest_joltage)

    return total_joltage

assert main("example.txt", part=1) == 357
assert main("input.txt", part=1) == 17343
```

```python
def parse_input(input):
    with open(input) as file_in:
        ranges_str, ingredients = file_in.read().split("\n\n")

    ranges_str = ranges_str.splitlines()
    ranges = []
    for rg_str in ranges_str:
        down, up = rg_str.split("-")
        ranges.append(range(int(down), int(up)+1))
    ranges = sorted(ranges, key = lambda x: x.start)

    ingredients = sorted(list(map(int, ingredients.splitlines())))

    return ingredients, ranges
```

```python
def split_ranges_union(range1, range2):
    if range2.start < range1.start or range1.start == range2.start and range1.stop < range2.stop:
        range1, range2 = range2, range1

    if range1.start <= range2.start and range2.stop <= range1.stop:
        return [range1]
    elif range1.stop >= range2.start and range2.stop >= range1.stop:
        return [range(range1.start, range2.stop)]
    elif range1.start < range2.start < range1.stop:
        return [range(range1.start, range2.start), range(range2.start, range1.stop), range(range1.stop, range2.stop)]
    else:
        return []
```

```python
def main(input, part):
    ingredients, ranges = parse_input(input)

    if part == 1:
        n_fresh_ingredients = 0
        for id in ingredients:
            for rg in ranges:
                if rg.start <= id <= rg.stop-1:
                    n_fresh_ingredients += 1
                    break

    elif part == 2:
        ranges = sorted(ranges, key = lambda x: x.start)
        ranges_final = []
        for j in range(len(ranges)):
            r_j = ranges[j]
            if j == 0:
                r_i = r_j
                continue
            union_r_i_r_j = split_ranges_union(r_i, r_j)
            if union_r_i_r_j:
                if len(union_r_i_r_j) == 1:
                    r_i = union_r_i_r_j[0]
                if len(union_r_i_r_j) == 3:
                    ranges_final.append(union_r_i_r_j[0])
                    ranges_final.append(union_r_i_r_j[1])
                    r_i = union_r_i_r_j[2]
            else:
                ranges_final.append(r_i)
                r_i = r_j
            if j == len(ranges)-1:
                ranges_final.append(r_i)
        
        n_fresh_ingredients = 0
        for rg in ranges_final:
            n_fresh_ingredients += len(rg)

    return n_fresh_ingredients
```

```python
assert main("example.txt", part=1) == 3
main("input.txt", part=1)
```

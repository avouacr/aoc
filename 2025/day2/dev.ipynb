```python
def main1(input):

    with open(input) as file_in:
        ranges = file_in.read().split(",")

    sum_invalid_ids = 0
    for range_ids in ranges:
        r_low, r_high = range_ids.split("-")
        for candidate in range(int(r_low), int(r_high)+1):
            candidate_str = str(candidate)
            if len(candidate_str) % 2 == 0:
                n1, n2 = candidate_str[:len(candidate_str)//2], candidate_str[len(candidate_str)//2:]
                if n1 == n2:
                    sum_invalid_ids += int(candidate_str)

    return sum_invalid_ids

assert main1("example.txt") == 1227775554
assert main1("input.txt") == 41294979841
```

```python
def main2(input):

    with open(input) as file_in:
        ranges = file_in.read().split(",")

    sum_invalid_ids = 0
    for range_ids in ranges:
        r_low, r_high = range_ids.split("-")
        range_invalid_ids = set()
        for candidate in range(int(r_low), int(r_high)+1):
            candidate_str = str(candidate)
            len_candidate = len(candidate_str)
            for size in range(1, len_candidate):
                if len_candidate % size == 0:
                    parts = set(candidate_str[i:i+size] for i in range(0, len(candidate_str), size))
                    if len(parts) == 1:
                        range_invalid_ids.add(candidate)
        sum_invalid_ids += sum(range_invalid_ids)

    return sum_invalid_ids

assert main2("example.txt") == 4174379265
assert main2("input.txt") == 66500947346
```

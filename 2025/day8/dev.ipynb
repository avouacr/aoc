```python
from collections import Counter
import logging
import sys
```

```python
def parse_input(path):
    with open(path) as file_in:
        data = file_in.read().splitlines()

    data = [tuple(map(int, row.split(","))) for row in data]
    return data
```

```python
def compute_dist(b1, b2):
    return ((b1[0]-b2[0])**2 + (b1[1]-b2[1])**2 + (b1[2]-b2[2])**2)**0.5
```

```python
def main1(input_path):

    if input_path == "example.txt":
        n_boxes_to_connect = 10
    elif input_path == "input.txt":
        n_boxes_to_connect = 1000

    boxes = parse_input(input_path)
    distances = []
    for i in range(len(boxes)):
        for j in range(i+1, len(boxes)):
            distances.append((boxes[i], boxes[j], compute_dist(boxes[i], boxes[j])))
    distances = sorted(distances, key=lambda x: x[2])[:n_boxes_to_connect]

    box2circuit = {}
    # print(box2circuit)
    n_connections = 0
    while n_connections < len(distances):
        b1 = distances[n_connections][0]
        b2 = distances[n_connections][1]
        # print(n_connections,  b1, b2)
        if b1 in box2circuit and b2 in box2circuit:
            if box2circuit[b1] != box2circuit[b2]:
                boxes_to_modify = [b for b in box2circuit if box2circuit[b] == box2circuit[b2]]
                for b in boxes_to_modify:
                    box2circuit[b] = box2circuit[b1]
        elif b1 in box2circuit and b2 not in box2circuit:
            box2circuit[b2] = box2circuit[b1]
        elif b2 in box2circuit and b1 not in box2circuit: 
            box2circuit[b1] = box2circuit[b2]
        elif b1 not in box2circuit and b2 not in box2circuit:
            if box2circuit == {}:
                new_circuit_id = 0
            else:
                new_circuit_id = max(box2circuit.values()) + 1
            box2circuit[b1] = new_circuit_id
            box2circuit[b2] = new_circuit_id
        n_connections += 1
        # print(n_connections, box2circuit)

    circuits = box2circuit.values()
    cnt = Counter(circuits)
    product = 1
    for circuit, size in cnt.most_common(3):
        product *= size
    return product
```

```python
assert main1("example.txt") == 40
main1("input.txt")
```

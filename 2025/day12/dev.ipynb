```python
import numpy as np
```

```python
def pattern_to_int_array(pattern):
    pattern_as_array = []
    for row in pattern.splitlines()[1:]:
        row_as_int = []
        for char in row:
            if char == "#":
                row_as_int.append(1)
            else:
                row_as_int.append(0)
        pattern_as_array.append(row_as_int)
    pattern_as_array = np.array(pattern_as_array)
    return pattern_as_array
```

```python
def parse_input(input_path):
    with open(input_path) as file_in:
        data = file_in.read().split("\n\n")

    shapes = {}
    for i, shp in enumerate(data[:-1]):
        pattern = shp.split(":")[1]
        pattern_int_array = pattern_to_int_array(pattern)
        shapes[i] = {}
        shapes[i]["pattern"] = pattern_int_array
        shapes[i]["area"] = pattern_int_array.sum()

    regions = {}
    for i, row in enumerate(data[-1].splitlines()):
        grid_size, shape_quantities = row.split(":")
        grid_size = tuple(map(int, grid_size.split("x")[::-1]))
        shape_quantities = tuple(map(int, shape_quantities.split(" ")[1:]))
        regions[i] = {}
        regions[i]["grid"] = grid_size
        regions[i]["quantities"] = shape_quantities

    return shapes, regions
```

```python
def get_max_capacity_used(region_index, regions, shapes):
    max_capacity_used = 0
    for i, qty in enumerate(regions[region_index]["quantities"]):
        max_capacity_used += qty * shapes[i]["area"]
    return max_capacity_used
```

```python
def main(input_file):
    shapes, regions = parse_input(input_file)
    n_regions_fit = 0
    for i in range(len(regions)):
        total_capacity_region = np.prod(regions[i]["grid"])
        total_area_shapes_to_fit = get_max_capacity_used(i, regions, shapes)
        if total_area_shapes_to_fit <= total_capacity_region:
            n_regions_fit += 1
    return n_regions_fit
```

```python
main("input.txt")
```
